```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Types of Plot

### Line Plots

```python
x = [1,2,3,4,5]
y = [10,20,25,30,40]
```

```python
plt.plot(x,y);
plt.title("Simple Line Plot")
plt.xlabel('X-Axis Label');
plt.ylabel('Y-Axis Label');
plt.grid(True)
```

## Bar Chart

```python
months = ['Jan', 'Feb', 'March', 'April']
revenue = [2000, 2500, 3000, 4000]
```

```python
plt.figure(figsize=(10,6))
plt.bar(months, revenue);
plt.title("Monthly Revenue Bar Chart")
plt.xlabel('Months');
plt.ylabel('Revenue');
```

## Histogram

```python
marks50 = np.random.randint(0, 100, 50)
marks50
```

```python
plt.hist(marks50);
plt.title("Marks Histogram")
plt.xlabel('Marks');
plt.ylabel('No. of Students');
```
